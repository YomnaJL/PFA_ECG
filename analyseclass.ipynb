```python
import os
import pandas as pd

# Définir le dossier contenant les images
image_folder = "image_classification"

# Lister toutes les images disponibles dans le dossier
available_images = set(os.listdir(image_folder))

# Extraire uniquement le nom du fichier depuis 'filename_lr' et ajouter ".png"
df["filename_only"] = df["filename_lr"].apply(lambda x: os.path.basename(x) + ".png")

# Filtrer le DataFrame
df_filtered = df[df["filename_only"].isin(available_images)].drop(columns=["filename_only"])

# Afficher sous forme de tableau
print(df_filtered.to_string(index=False))  # Supprime l'index pour un affichage propre
```

```python
from tabulate import tabulate  # Pour un affichage joli dans la console

# Sélectionner uniquement les colonnes d'intérêt
cols = ['Myocardial Infarction (MI)', 'ST/T Change (STTC)', 'Conduction Disturbance (CD)', 
        'Hypertrophy (HYP)', 'Normal (NORM)']

# Compter le nombre de fois où chaque colonne a la valeur 1 dans le DataFrame filtré
class_counts_filtered = (df_filtered[cols] == 1).sum()

# Afficher la distribution sous forme de tableau
print("\nDistribution des classes après filtrage des images disponibles :")
print(tabulate(class_counts_filtered.to_frame(), headers=["Classe", "Nombre d'occurrences"], tablefmt="grid"))
```

```python
import pandas as pd

# Colonnes à équilibrer
cols_to_balance = ['Conduction Disturbance (CD)', 'Hypertrophy (HYP)']

# Supprimer les lignes déjà présentes dans df_filtered
df_remaining = df[~df["filename_lr"].isin(df_filtered["filename_lr"])].copy()

# Liste pour stocker les lignes à ajouter
rows_to_add = []

# Ajouter jusqu'à 1000 lignes pour chaque classe == 1
for col in cols_to_balance:
    candidates = df_remaining[df_remaining[col] == 1]
    to_add = candidates.sample(n=min(1000, len(candidates)), random_state=42)
    rows_to_add.append(to_add)
    df_remaining = df_remaining.drop(to_add.index)

# Fusionner avec df_filtered (éviter doublons sur filename_lr)
added_rows = pd.concat(rows_to_add)
df_combined = pd.concat([df_filtered, added_rows]).drop_duplicates(subset="filename_lr").reset_index(drop=True)

# ✅ Afficher le tableau complet
print("✅ Tableau combiné avec toutes les colonnes :")
display(df_combined)
```
